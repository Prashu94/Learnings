```python
#Question 4
X = np.array([1,2,3,4])
Y = np.array([20,21,20.5,20.8])
print (X,Y)
```

```python
#Draw a simple plot
plt.plot(X,Y)
plt.show()
```

```python
#Configure Line and markers in simple plot
plt.plot(X,Y,linewidth = 2,color='blue',linestyle='dashed',marker='o',markersize=12)
plt.show()
```

```python
#Configure the axes
plt.plot(X,Y,linewidth = 2,color='blue',linestyle='dashed',marker='o',markersize=12)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Give title of the graph
plt.plot(X,Y,linewidth = 2,color='blue',linestyle='dashed',marker='o',markersize=12)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Simple Lined and Marked Plot")
plt.show()
```

```python
#Give Error Bar
Y_error= np.array([0.12,0.13,0.2,0.1])
plt.plot(X,Y,linewidth = 2,color='blue',linestyle='dashed',marker='o',markersize=12)
plt.plot(X,Y_error,linewidth = 2,color='red',linestyle='dashed',marker='o',markersize=12)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Error Bar")
plt.show()
```

```python
#Define width and height as figsize=(4,5) and adjust plot dpi =100
plt.plot(X,Y,linewidth = 2,color='blue',linestyle='dashed',marker='o',markersize=12)
plt.figure(figsize = (4,5),dpi=100)
plt.show()
```

```python
#Creatig a dataframe manually

data = {
    'first_name':pd.Series(['Jason','Molly','Tina','Jake','Amy']),
    'last_name':pd.Series(['Miller','Jacobson','Ali','Milner','Cooze']),
    'female':pd.Series([0,1,1,0,1]),
    'age':pd.Series([42,52,36,24,73]),
    'preTestScore':pd.Series([4,24,31,2,3]),
    'postTestScore':pd.Series([25,94,57,62,70])
}

data1= pd.DataFrame(data)
data1

#Draw a scatterplot of pre test scroe and post test score
plt.scatter(data1['preTestScore'],data1['postTestScore'])
plt.xlabel("Pre-TestScore")
plt.ylabel("Post-TestScore")
plt.title("ScatterPlot of pre and post test score")
plt.show()
```

```python
#Draw a Scatterplot from the data in question 9 of preTestScore and 
#postTestScore with the size = 300 and the color determined by sex

plt.scatter(data1['preTestScore'],data1['postTestScore'],c=data1['female'],cmap=plt.cm.autumn)
plt.xlabel("Pre-TestScore")
plt.ylabel("Post-TestScore")
plt.title("ScatterPlot of pre and post test score")
plt.show()
```
